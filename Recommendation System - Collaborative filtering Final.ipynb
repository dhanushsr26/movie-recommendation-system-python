## Data Preprocessing
- We planned to merge two dataset on `metadata-id` and `ratings-movieId`
- After going through the info and summary we have found out that `metaadata-id` column is actually **tmdb_id** which is not same as `ratings-movieId`
- We explored the kaggle dataset and found a solution. There is a **link.csv** file which has mainly 3 attributes `movieId,imdbid,tmdbid`
- So we are going to make new movieId column on metadata using merge function mapping on tmbdid from `metadata` and movieId from `link`

```python
link=pd.read_csv('links.csv')
link.info()
```

##### Converting `id` from string to integer
- From the `metadata.info` we found that the `id` is in string so we need to convert it into numeric format to map the two datasets 

```python
metadata['id'] = pd.to_numeric(metadata['id'], errors='coerce')
metadata = metadata.dropna(subset=['id'])
metadata['id'] = metadata['id'].astype(int)
```

```python
metadata = pd.merge(link, metadata, left_on='tmdbId', right_on='id')
```

```python
metadata.info()
```

##### Extraction of data
- We only need the **orginal title** and **movieid** from `metadata`
- Combine with `ratings` 

```python
metadata = metadata[['movieId', 'original_title']]
```

```python
data=pd.merge(rating,metadata,on='movieId')
```

##### Extracting Popular Movies

```python
top_movies = data['original_title'].value_counts().head(1500).index
data = data[data['original_title'].isin(top_movies)]
```

## User-item Matrix
- Took a sample of 1 million.
- Created a pivot table.
- 

```python
data
```

```python
user_movie_matrix = data.pivot_table(
    index='userId',
    columns='original_title',
    values='rating'
)
```

```python
user_movie_matrix
```

```python
user_movie_matrix.notna().sum().sum()
```

##### Calculate cosine similarity
- Replacing `Nan` in the matrix with *0*
- Made a matrix using cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# Replace NaN with 0
matrix_filled = user_movie_matrix.fillna(0)

similarity = cosine_similarity(matrix_filled.T)

import pandas as pd
similarity_df = pd.DataFrame(
    similarity,
    index=user_movie_matrix.columns,
    columns=user_movie_matrix.columns
)
```

```python
similarity_df
```

##### Sanity Checking of recommendation system
- We used item-item collaborative filtering with cosine similarity. Movies are represented as vectors of user ratings. Similarity is computed based on user rating patterns, and recommendations are generated by averaging similarity scores across selected movies.

```python
similarity_df["Toy Story"].sort_values(ascending=False).head(10)
```

##### Making it more user friendly
` Constructed for multiple movie inputs

```python
def recommend_safe(movie_list, n=5):
    valid_movies = []
    
    for m in movie_list:
        matches = [x for x in similarity_df.columns if m.lower() in x.lower()]
        if matches:
            valid_movies.append(matches[0])

    if not valid_movies:
        return []   # ← always return list

    scores = similarity_df[valid_movies].mean(axis=1)
    scores = scores.sort_values(ascending=False)
    scores = scores.drop(valid_movies)

    return list(scores.head(n).index)
```

```python
recommend_safe([
  "Random Unknown Movie"
])
```

```python
def recommend(movie_list, n=5):
    recs = recommend_safe(movie_list, n)
    print("Recommended movies:\n")
    for i, m in enumerate(recs, 1):
        print(f"{i}. {m}")
```

```python
recommend([
  "Mission"
])
```

```python
recommend(["toy",'mission'],n=10)
```
